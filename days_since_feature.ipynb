```python
import pandas as pd
```

```python
v5 = pd.read_csv('./assets/skunkworks_data_v5.csv')
```

```python
v5.head(10)
```

```python
v5['Date'] = pd.to_datetime(v5['Date'])
```

```python
v5.head()
```

```python
campaign_first_dates = v5.groupby('Campaign')['Date'].min()
```

```python
print(campaign_first_dates)
```

```python
v5['Days_Since_Campaign_Start'] = v5.apply(
    lambda row: (row['Date'] - campaign_first_dates[row['Campaign']]).days,
    axis=1
)
```

```python
print(v5[['Date', 'Campaign', 'Days_Since_Campaign_Start']].head(10))
```

```python
v5.head()
```

```python
v5.to_csv('./assets/skunkworks_data_v6.csv', index=False)
```
